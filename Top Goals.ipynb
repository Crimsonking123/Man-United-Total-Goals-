```python
#This script scrapes goal data from transfermarketfor squads
import selenium
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.ui import WebDriverWait
import time
browser = webdriver.Firefox()
```

```python
#The function returns a season name for a year inputted
def create_seasons(a):
    season_str = str(a) + "/" + str(a+1)
    return season_str
```

```python
import pandas as pd
columns = ["Players","Season","Goals","Minutes"]
Players_1,Season_1,Goals_1,Minutes_1 = [],[],[],[]

#This finds the rows of the table for each page containing a player
def rawdata(current):
    link = "https://www.transfermarkt.com/manchester-united/leistungsdaten/verein/985/plus/0?reldata=%26"
    link += str(current)
    browser.get(link)
    time.sleep(2)
    test = browser.find_elements_by_xpath("//tr[@class='even'] | //tr[@class = 'odd']")
##While iterating through the elements - cleanup os required. First we split the rows into different lines by /n    
    for i in test:
        data = i.text.splitlines()  
        last = len(data) - 1
#This appends the split line containing the player name
        Players_1.append(data[1])
#This appends the generated season from the inputted variable
        Season_1.append(create_seasons(current))
#This appends zero as the default value for goals and minutes for players who didn't play in a season
        if data[last].find("during this season") != -1:
            Goals_1.append(0)
            Minutes_1.append(0)
#This aplits the last line of the data variable which contains stats into a list by whitespace
        else:
            stats = data[last].split(" ")
            last = len(stats) - 1
            goals_val = stats[last-1]
            mins_val = stats[last]
#The goals value is replaced by zero is NaN and the minutes value is replaced by zero if NaN
            if goals_val == "-":
                Goals_1.append(0)
                mins_val = mins_val.replace("'","")
                if mins_val == "-":
                    Minutes_1.append(0)
                else:
                    Minutes_1.append(float(mins_val))
            else:
                goals_val = goals_val.replace("'","")
                mins_val = mins_val.replace("'","")
                Goals_1.append(int(goals_val))
                Minutes_1.append(float(mins_val))
```

```python
#Create a dataframe 
current_year = 1962
while current_year != 2021:
    rawdata(current_year)
    current_year += 1
dict_1 = {}
category = [Players_1,Season_1,Goals_1,Minutes_1]
for i in range(len(columns)):
    dict_1[columns[i]] = category[i]
Table_1 = pd.DataFrame(dict_1,columns=columns)
```

```python
#This returns two lists containing the unique player names and seasons
Unique_names = Table_1["Players"].unique()
Unique_seasons = Table_1["Season"].unique()




```

```python

Reordered = pd.DataFrame(columns=["Players","Season","Goals"])
#This generates an individual data frame for each player in the tab;e with a matching unique name
for i,ele in enumerate(Unique_names):
    Unique_frame = Table_1[Table_1["Players"] == ele].reset_index(drop=True)
    del Unique_frame["Minutes"]
    season_list = list(Unique_frame["Season"])
#This generates an index from the numerical value of the first year of the seasons
    index = [int(x[0:4]) for x in season_list]
    Unique_frame.index = index
#If the season isn't in our unique dataframe, we insert it by the index generated from its first year into our unique dataframe
    for j,item in enumerate(Unique_seasons):
        doesexist = Unique_frame[Unique_frame["Season"] == item]
        if len(doesexist) == 0:
            Unique_frame.loc[int(item[0:4])] = ele, item, 0
            Unique_frame = Unique_frame.sort_index()
#We merge our empty Reordered dataframe with the current unique player dataframe
    result = [Reordered,Unique_frame]
    result = pd.concat(result)
    Reordered = result.reset_index(drop=True)

    


```

```python
final_dict = {}
final_dict["Season"] = Unique_seasons.tolist()

#An empty cumulative goals column is created
Reordered["Cum_goals"] = ["" for x in range(len(Reordered["Goals"]))]
#We create a dataframe from rows matching our unique player name
for i,ele in enumerate(Unique_names):
    final_list = []
    player_frame = Reordered[Reordered["Players"] == ele]
    length = len(player_frame)
#The first index where the player appears can then be obtained and cumulative goals can be matched to the starting goals row
    pos = player_frame.index[0]
    Reordered.at[pos,"Cum_goals"] = Reordered.at[pos,"Goals"]
    final_list.append(Reordered.at[pos,"Cum_goals"])
#Afterwards cumulative goals is the sum of the last cumulative goal value and the current goal value
    for i in range(length-1):
        pos += 1
        Reordered.at[pos,"Cum_goals"] = Reordered.at[pos,"Goals"] + Reordered.at[pos-1,"Cum_goals"]
        final_list.append(Reordered.at[pos,"Cum_goals"])
#Ashley Grimes is skipped due to wrong formatting compared to the rest of the club by transfermarket
    if ele == "Ashley Grimes":
        pass
    else:
        final_dict[ele] = final_list
    

```

```python
Unique_names = Unique_names.tolist()
Unique_names.remove("Ashley Grimes")


```

```python
#Finally a dataframe with index of Seasons and player columns is made
final_dataframe = pd.DataFrame(final_dict)
final_dataframe.set_index("Season",inplace=True)
final_dataframe
```
